```python
import pandas as pd

df = pd.read_csv("greendb - hotel.csv")
df
```

```python
import re
import os
#SQL COMMAND FOR IMPORTING DATA INTO BUSINESS TABLE
lst = list()
path = "assets/press-package/"
for index, rows in df.iterrows():
    pdf = ""
    for filename in os.listdir(path):
        filename = re.findall('(^.*).pdf',filename)
        if rows.Name == filename[0]:
            pdf = filename[0]
    if len(pdf) != 0:
        lst.append("INSERT INTO BUSINESS VALUES (" +str(index)+",\"" +rows.Category +"\",\"" +rows.Name +"\",\"" +rows.Street+"\",\""+
              rows.City+"\",\""+str(rows.ZIP)+"\",\""+rows.Number+"\",\""+rows.Description+"\",\""+rows.Name+".png\",\""+
               pdf+".pdf\");")
    if len(pdf) == 0:
        print(rows.Name,"vs.",pdf)
for l in lst:
    print(l)
```
